```python
def simple_Linear_Regression(input_feature,output):
    numerator=np.mean(input_feature*output)-(np.mean(input_feature))*(np.mean(output))
    denominator=np.mean(input_feature**2)-(np.mean(input_feature))*(np.mean(input_feature))
    slope=numerator/denominator
    a=np.mean(output)
    b=slope*np.mean(input_feature)
    intercept=a-b
    return slope,intercept
```

```python
input_feature=train_data['sqft_living']
output=train_data['price']
slope,intercept=simple_Linear_Regression(input_feature,output)
print(slope,intercept)
```

```python
def get_regression_predictions(predict,intercept,slope):
    output=predict*slope+intercept
    print(output)
```

```python
predict=2650
get_regression_predictions(predict,intercept,slope)
```

```python
def get_residual_sum_of_squares(predict,output,intercept,slope):
    rss=sum((output-((predict*slope)+intercept))**2)
    print(rss)
```

```python
get_residual_sum_of_squares(train_data['sqft_living'],train_data['price'],intercept,slope)
```

```python
def inverse_regression_predictions(output,intercept,slope):
    input_data=(output-intercept)/slope
    print(input_data)
```

```python
cost=800000
inverse_regression_predictions(cost,intercept,slope)
```
